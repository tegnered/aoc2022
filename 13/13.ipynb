```python
comp1 = []
comp2 = []
for i in range(0,len(inp),3):
    comp1.append(eval(inp[i]))
    comp2.append(eval(inp[i+1]))
```

```python
def compare(c1,c2):
    print(f'Compare {c1} and {c2}')
    if type(c1) == int and type(c2) == int:
        if c1 < c2:
            return True
        elif c1 > c2:
            return False
    elif type(c1) == list and type(c2) == list:
        for s1,s2 in zip(c1,c2):
            res = compare(s1,s2)
            if res == True:
                return True
            elif res == False:
                return False
        if len(c1) < len(c2):
            return True
        elif len(c1) > len(c2):
            return False
    elif type(c1) == int and type(c2) == list:
        return compare([c1],c2)
    elif type(c1) == list and type(c2) == int:
        return compare(c1,[c2])
```

```python
idxs = []
for i in range(len(comp1)):
    if compare(comp1[i],comp2[i]):
        idxs.append(i+1)
```

```python
divs = [[[2]], [[6]]]
arr = comp1 + comp2 + divs
def bubbleSort(arr):
    n = len(arr)
    # optimize code, so if the array is already sorted, it doesn't need
    # to go through the entire process
    swapped = False
    # Traverse through all array elements
    for i in range(n-1):
        # range(n) also work but outer loop will
        # repeat one time more than needed.
        # Last i elements are already in place
        for j in range(0, n-i-1):
 
            # traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if compare(arr[j],arr[j + 1]):
                swapped = True
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
         
        if not swapped:
            # if we haven't needed to make a single swap, we
            # can just exit the main loop.
            return
```

```python
bubbleSort(arr)
```
